```python
import pandas as pd 
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_excel(r"C:\Users\HO SONG TIN\Downloads\tkud.xlsx")
```

```python
df.drop('Nhóm chúng em/mình rất cảm ơn vì sự giúp đỡ của anh/chị/bạn. Anh/chị/bạn hãy để lại link khảo sát của nhóm mình dưới đây (nếu có) để chúng em/mình hỗ trợ khảo sát nha.', axis=1, inplace=True)
```

```python
df.columns = ['time', 'khoa', 'năm','phone using purpose', 'use phone before sleep', 'avg time using phone', 'using phone before sleep', 'first section ratio', 'lost focus?', 'sleep effiecient', 'opinion1', 'opinion2', 'opinion3', 'opinion4', 'opinion5', 'opinion6']
```

```python
df
```

```python
df['năm'] = df['năm'].map(str)

df_c = df.groupby(['khoa']).size().reset_index()
df_n = df.groupby(['năm']).size().reset_index()
```

```python
df_n
```

```python
fig = plt.figure(figsize=(27,6))
colors = sns.color_palette('pastel')

ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)


ax0.pie(df_c[0], labels=df_c['khoa'], labeldistance=1.1, autopct='%1.0f%%', colors=colors)
ax0.axis('equal')


ax1.bar(df_n['năm'], df_n[0], color=colors)
ax1.set_xlabel('Năm', fontsize=15)
ax1.set_ylabel('Số lượng', fontsize=15)

plt.suptitle('Sinh viên uel tham gia khảo sát', fontsize=35, y=1.05, fontname="Times New Roman", weight='bold');
```

```python
fig = plt.figure(figsize=(10,6))

df_p = df.groupby(['phone using purpose']).size().reset_index()
df_p.loc[len(df_p.index)] = ['Không', 0]
plt.pie(df_p[0], colors=sns.color_palette('pastel'), autopct='%1.0f%%', pctdistance=0)
plt.title('Tỉ lệ sử dụng điện thoại trong 1 tiếng trước khi đi ngủ', fontsize=15, y=1.05)
plt.axis('equal')
plt.legend(labels=df_p['phone using purpose'], loc='upper left');
```

```python
dic = {'num': []}
for item in df['sleep effiecient']:
    if item == 'Tốt':
        dic['num'].append(4)
    elif item == 'Không ổn':
        dic['num'].append(1)
    elif item == 'Tạm chấp nhận được':
        dic['num'].append(2)
    elif item == 'Vừa đủ':
        dic['num'].append(3)
    elif item == "Rất tốt":
        dic['num'].append(5)
lst = dic['num']
df['num'] = lst
```

```python
df.rename(columns = {'sleep effiecient': "Phân loại", 'using phone before sleep':'Thời gian'}, inplace=True)
df
```

```python
fig = plt.figure(figsize=(10,6))


sns.histplot(data=df, x='num', hue='Phân loại', bins=5)
plt.xlabel('Chất lượng giấc ngủ')
plt.ylabel('Tần số')
plt.tick_params(labelbottom = False, bottom = False)
```

```python
lst = []
for item in df['Thời gian']:
    if item == '0 phút - 10phút':
        lst.append(1)
    if item == '11 phút - 20 phút':
        lst.append(2)
    if item == '21 phút - 30 phút':
        lst.append(3)
    if item == '31 phút - 40 phút':
        lst.append(4)
    if item == '40 phút trở lên':
        lst.append(5)
df['Thời gian sử dụng điện thoại trước khi ngủ'] = lst
```

```python
new_df = df.sort_values(by='Thời gian sử dụng điện thoại trước khi ngủ')
fig = plt.figure(figsize=(10,6))

sns.histplot(data=new_df, x='Thời gian sử dụng điện thoại trước khi ngủ', hue='Thời gian', bins=5)
plt.tick_params(labelbottom = False, bottom = False)
plt.xlabel('Thời gian sử dụng điện thoại trước khi ngủ')
plt.ylabel('Tần số');
```
